# Timeseries LSTM Training

Train a 3-layer LSTM (128 hidden units) on fixation sequences

```python
from pathlib import Path

import pandas as pd

from src.dl.training import (
    DEFAULT_MAX_LENGTH,
    DEFAULT_TIMESERIES_FEATURES,
    LSTM_HIDDEN_SIZE,
    LSTM_NUM_LAYERS,
    run_training_battery,
)

DATASETS_DIR = Path("data")
SPLITS_DIR = Path("extensive_features") / "splits"
RESULTS_DIR = Path("results") / "dl"
RESULTS_FILE = RESULTS_DIR / "lstm_results.csv"

MAX_EPOCHS = 50
BATCH_SIZE = 32
TEST_MODE = False
TEST_MAX_SAMPLES = 100
SKIP_EXISTING = True

RESULTS_DIR.mkdir(parents=True, exist_ok=True)

print("Configuration:")
print(f"  datasets_dir: {DATASETS_DIR}")
print(f"  splits: {SPLITS_DIR}")
print(f"  results: {RESULTS_FILE}")
print(f"  LSTM: {LSTM_NUM_LAYERS} layers x {LSTM_HIDDEN_SIZE} units")
print(f"  sequence features: {DEFAULT_TIMESERIES_FEATURES}")
print(f"  max_length: {DEFAULT_MAX_LENGTH}")
print(f"  max_epochs: {MAX_EPOCHS}, batch_size: {BATCH_SIZE}")
print(f"  test_mode: {TEST_MODE}")
```

```python
results_df = run_training_battery(
    splits_dir=SPLITS_DIR,
    results_file=RESULTS_FILE,
    datasets_dir=DATASETS_DIR,
    max_epochs=MAX_EPOCHS,
    batch_size=BATCH_SIZE,
    timeseries_features=DEFAULT_TIMESERIES_FEATURES,
    max_length=DEFAULT_MAX_LENGTH,
    skip_existing=SKIP_EXISTING,
    test_mode=TEST_MODE,
    test_max_samples=TEST_MAX_SAMPLES,
)

if results_df.empty and RESULTS_FILE.exists():
    results_df = pd.read_csv(RESULTS_FILE)

print(f"\nDone. {len(results_df)} rows in {RESULTS_FILE}")
```
